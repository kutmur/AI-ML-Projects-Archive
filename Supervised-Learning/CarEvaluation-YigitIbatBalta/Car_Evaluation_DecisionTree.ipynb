```python
X = df.drop("class" , axis = 1)
y = df["class"]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.25 , random_state = 42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.columns
```

```python
categoricall_cols = ["Buying" , "maint" , "lug_boot" , "safety"]
```

```python
df["safety"].value_counts()
```

```python
numerical_cols = ["doors" , "persons"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[
    ["low","med","high","vhigh"],
    ["low","med","high","vhigh"],
    ["small","med","big"],
    ["low","med","high"]
])
```

```python
X_train[categoricall_cols] = ordinal_encoder.fit_transform(X_train[categoricall_cols])
```

```python
X_test[categoricall_cols] = ordinal_encoder.transform(X_test[categoricall_cols])
```

```python
X_test
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "criterion": ["gini" , "entropy" , "log_loss"],
    "splitter" : ["best","random"],
    "max_features" : ["auto","sqrt","log2"]
}
```

```python
grid = GridSearchCV(estimator=model , param_grid = param_grid , n_jobs = -1 , scoring = "accuracy")
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred = grid.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
